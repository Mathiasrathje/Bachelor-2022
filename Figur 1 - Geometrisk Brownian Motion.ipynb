## GMB - Figur 1

```python
import matplotlib.pyplot as plt
import numpy as np


T = 2
mu = 1
sigma = 0.1
sigma1 = 0.4
sigma2 = 0.7
S0 = 1
dt = 0.01
N = round(T/dt)
t = np.linspace(0, T, N)

W = np.random.standard_normal(size = N) #Noter at vi har benyttet samme W#
W = np.cumsum(W)*np.sqrt(dt) ### standard brownian motion ###

X = (mu-0.5*sigma**2)*t + sigma*W
X1 = (mu-0.5*sigma1**2)*t + sigma1*W 
X2 = (mu-0.5*sigma2**2)*t + sigma2*W 

S = S0*np.exp(X)### geometric brownian motion ###
S1 = S0*np.exp(X1)
S2 = S0*np.exp(X2)
Expected_value = 1*np.exp(mu*t)

plt.plot(t, S, label='sigma = 0.1')
plt.plot(t,S1, label='sigma = 0.4')
plt.plot(t,S2, label='sigma = 0.7')
plt.plot(t,Expected_value,'Black', label='Expected_value')
plt.legend()
plt.xlabel('t',fontsize=12)
plt.ylabel('X_t',fontsize=12)
plt.show()
```
